```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x):     #начальная функция
    return 2*x[0]**2 + 3*x[1]**2 - 2*np.sin((x[0]-x[1])/2)+x[1]
def f_x1(x):  #производная по x1
    return 4*x[0] - np.cos((x[0]-x[1])/2)
def f_x2(x):  #производная по x2
    return 6*x[1] + np.cos((x[0]-x[1])/2) + 1

def grad_met(x,eps):
    a=1
    while(a==1):  #метод наискорейшего спуска
        h=[0,0]
        h[0]=-f_x1(x)  #направление по x1
        h[1]=-f_x2(x)  #направление по x2
        s=2
        g0=1000000
        while True:  #вычисление наилучшего альфа( argminf(xk - alf * f`(xk)) )
            xg=[0,0]
            xg[0] = x[0]-s*f_x1(x)
            xg[1] = x[1]-s*f_x2(x)
            g = f(xg)
            if(g<g0):
                alf=s
                s=s/2
                g0=g
            else:
                break
        w=f(x) 
        for i in range(2): #получаем координаты нового х
            x[i]=x[i]+alf*h[i]
        if(abs(f(x)-w)<eps):    #условие остановки изфункции с координатами нового икса вычитаем начальную
            a=0
    print('Минимум достигается в точке (',x[0],';',x[1],') и равен ',f(x))
```

```python
eps=0.01
x0=[0,0] #начальное положение
grad_met(x0,eps)
```

```python
x = np.arange (-10, 10, 0.1)
y = np.arange (-10, 10, 0.1)
xgrid, ygrid = np.meshgrid(x, y)
zgrid = 2*xgrid**2 + 3*ygrid**2 - 2*np.sin((xgrid-ygrid)/2)+ygrid 

fig1 = plt.figure()
axes1 = Axes3D(fig1)
axes1.plot_surface(xgrid, ygrid, zgrid)

plt.show()
```
