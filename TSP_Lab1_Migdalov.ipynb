```python
import numpy as np
from numpy import random
import random
from matplotlib import pyplot as plt
import math
import scipy.stats as st
from scipy.stats import kstest
import pandas as pd
```

### Число траекторий равно 1000, время Т

```python
#Начальные данные 
p = 0
l = 0
a = 0
b = 15
mu = (a + b) / 2
T = 100 * mu
n = 1000    #для построения 100 траекторий приравнять n к 100
z = np.zeros(n)

#Формирование траекторий процесса и массива z, нужного для проверки ЦПТ, и параметров, нужных для проверки ЗБЧ
for i in range(n):
    s = [0]
    while s[-1] < T:        #для точки Т/2, поменять Т на Т/2
        s.append(s[-1] + np.random.uniform(a,b))
    k = len(s)
    e = (k - T/mu)**2
    l += k
    p += e
    z[i] = (k - T/mu) 
    plt.plot(s, range(len(s)))
plt.show()
z = z / math.sqrt(p/n)


#Рисуем значения z
Norm = np.zeros(n)
for i in range(n):
    Norm[i] = np.random.normal(0, 1)
    plt.plot(range(len(z)), z)
plt.plot(Norm, color = 'red')
plt.show()

#Поверка ЗБЧ    
print("H(T) с крышкой для n траекторий (n = 1000) = ", l / n, "\n", 
      "H(T) = T/mu = ", T/mu)

#Гистограмма z и плотность нормального распределения
z.sort()
fit1 = st.norm.pdf(z, np.mean(z), np.std(z)) #плотность нормального распределнеия 
fit2 = st.norm.pdf(z, 0, 1) #плотность стандартного нормального распределнеия
plt.plot(z, fit1, 'r--')
plt.plot(z, fit2, 'g--')
plt.hist(z, bins = 20, normed = True, color = 'lightblue')
plt.show()


#Проверка на нормальность по критерию Колмагорова
print(kstest(z,'norm'))  
```
